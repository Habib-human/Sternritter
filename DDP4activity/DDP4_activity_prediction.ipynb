```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('Dipeptidyl-peptidase IV')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
df_int = pd.read_csv('DDP4_bioactivity_data_preprocessed.csv')
```

```python
bioactivity_threshold = []
for i in df_int.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df_label = pd.concat([df_int, bioactivity_class], axis=1)
df_label
```

```python
df_label.to_csv('DDP4_bioactivity_data_curated.csv', index=False)
```

```python
df_curated = pd.read_csv('DDP4_bioactivity_data_curated.csv')
df_curated
```

```python
df_no_smiles = df_curated.drop(columns='canonical_smiles')
```

```python
smiles = []

for i in df_curated.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
df_clean_smiles = pd.concat([df_no_smiles,smiles], axis=1)
df_clean_smiles
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df_clean_smiles.canonical_smiles)
df_lipinski
```

```python
df_curated
```

```python
df_combined = pd.concat([df_curated,df_lipinski], axis=1)
df_combined
```

```python
import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)
        
    return x
```

```python
df_combined.standard_value.describe()
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)
        
    return x
```

```python
df_norm = norm_value(df_combined)
df_norm
```

```python
df_norm.standard_value_norm.describe()
```

```python
df_final = pIC50(df_norm)
df_final
```

```python
df_final.pIC50.describe()
```
